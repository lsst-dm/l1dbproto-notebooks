```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
ras = np.arange(0., 2.*np.pi, 0.1)
count = len(ras)
```

```python
fig, ax = plt.subplots(subplot_kw={'projection': 'polar'}, figsize=(15, 15))
ax.set_rmax(110)
for dec in np.arange(50., 55., 10.):
    decs = np.full((count,), dec)
    ax.scatter(ras, decs)
```
